### Lesing av fil

Vi kan lese av filer vha. funksjonen _loadtxt_:
data = loadtxt('filnavn',skiprows=1,delimiter=',',dtype=str,usecols=(1,2,4)) 

funksjonen _mean_ brukes for å finne gjennomsnittet av lista "hoyde"
funksjonen _std_ (standard diviation) brukes for å finne standard avviket, det gir altså et utrykk for usikkerheten til målingene i lista "hoyde"

```python
from pylab import *

data = loadtxt("datatekst.txt",skiprows=1,delimiter=",")

nr = data[:,0]
hoyde = data[:,1]

plot(nr,hoyde)

snitt = mean(hoyde)
avvik = std(hoyde)   #standard diviation


print("snitt:", snitt, "Standardavvik:", avvik)
```

## regresjon

Regresjon betyr å finne en funksjon som er best mulig tilpasset alle punktene i et datasett. 

Bruker funksjonen _plyfit_ til å lage en regresjons analyse, polyfit(x,y,grad), den lager en array med  a,b,c... verdiene til regresjons funksjonen

```python
from pylab import *
from sklearn.metrics import r2_score

T = array([0,20,40,60,80,100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])



# Regresjon av NaCl
grad = 2
reg_NaCl = polyfit(T,sol_NaCl,grad)

x = linspace(0,150,10000)
a = reg_NaCl[0]
b = reg_NaCl[1]
c = reg_NaCl[2]
y = a*x**2 + b*x + c
y1_R2 = a*T**2 + b*T + c

R2 = r2_score(sol_NaCl, y1_R2)

scatter(T,sol_NaCl)
plot(x,y,color='red')
show()
print(R2)


# Regresjon av NH3
grad2 = 3
reg_NH3 = polyfit(T,sol_NH3,grad2)
A = reg_NH3[0]
B = reg_NH3[1]
C = reg_NH3[2]
D = reg_NH3[3]
Y = A*x**3 + B*x**2 + C*x + D
y2_R2 = A*T**3 + B*T**2 + C*T + D

R2_2 = r2_score(sol_NH3, y2_R2)

scatter(T,sol_NH3)
plot(x,Y,color='red')
show()
print(R2_2)
```

## Statistisk analyse

#### Gjennomsnitt

Gjennomsnittet av $n$ målinger er definert som summen av alle målingene dividert på $n$:  

$$\bar{x}=\frac{1}{n}\sum_{i=1}^n x_i$$

#### Standaravvik

Standardavviket til $n$ målinger er definert slik: 

$$s= \sigma^2=\sqrt{\frac{1}{n}\sum_{i=1}^n (x_i-\bar{x})^2}$$
