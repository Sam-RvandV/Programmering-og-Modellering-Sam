# Numerisk metode - derivasjon


## Definisjoner

### Analytisk definisjon

$$f'(x) =\space \lim_{∆a\rightarrow0} \space \frac{f(x+∆x)-f(x)}{∆x}$$

### Nummerisk definisjon

#### Newtons kvotient

Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.

$f'(x)$ ≈ $\frac{f(x+h)-f(x)}{h}$

#### Newtons symmetriske kvotient

Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.
Denne gir enda litt bedre resultater enn den over.

$f'(x)$ ≈ $\frac{f(x+h)-f(x-h)}{2h}$


### Eksempel med bruk av Newtons kvotient

```python
from pylab import *

def f(x):
    return x**2

def fder(f,x,dx=1E-8):
    return (f(x+dx)-f(x))/dx

df = fder(f,1)

print(df)
```

# Derivasjon av diskrete data

### Formel

Bruker en tilnærming for den deriverte som vi skjenner fra matten som momentan vekst 
(med noe avvik)

$d$ står for en liten (infinitesimal!) endring

$\frac{∆y}{∆x} = \frac{y2-y1}{x2-x1} \rightarrow \frac{dy}{dx}$

##### Diskret derivasjon kan ikke gjøres med analytiske metoder

```python
from pylab import *

t = [5,10,15,30,50,60]
s = [115, 190, 390, 600, 1000, 1200]
v = []
v.append(0)
a = []
a.append(0)

for i in range(len(s)-1):
    sder = (s[i+1] - s[i])/(t[i+1]-t[i])
    v.append(sder)
    
for k in range(len(v)-1):
    vder = (v[k+1] - v[k])/(t[k+1]-t[k])
    a.append(vder)

    
plot(t,s)
xlabel("tid")
ylabel("strekning som funksjon av tid")
show()

plot(t,v)
xlabel("tid")
ylabel("fart som funksjon av tid")
show()

plot(t,a)
xlabel("tid")
ylabel("aksellerasjon som funksjon av tid")
show()
```

## Testing

### Newtons kvotient

```python
from pylab import *

x = linspace(-10,10,21)

def f(x):
    return x**2

def fder(f, x, dx=1E-8):
    return (f(x+dx)-f(x))/dx

df = fder(f,x)

plot(x,f(x))
show()
plot(x, df)
show()
```

### Newtons symmetriske kvotient

```python
from pylab import *

x = linspace(-10,10,21)

def f(x):
    return x**2

def fder(f, x, h=0.1):
    return (f(x+h)-f(x-h))/(2*h)

df = fder(f,x)

plot(x,f(x))
show()
plot(x, df)
show()
```

### Later som at grafen $f(x)=x^2$ er diskrete data

```python
x = linspace(-10,10,21)

def f(x):
    return x**2

y = f(x)

dx = []

for i in range(len(x)-1):
    xder = (y[i+1] - y[i])/(x[i+1]-x[i])
    dx.append(xder)

print(len(dx))

plot(x,y)
show()
plot(x[0:20],dx)
show()
```
