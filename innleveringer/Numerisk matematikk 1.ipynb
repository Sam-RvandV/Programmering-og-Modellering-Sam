#### a) 

En sekant er en rett linje som kjærer i gjennom en kurve i minst to punkter. Det kan brukes til å bestemme gjennomsnittlig vekst, dette brukes også i den deriverte men da sier man at disse punktene går mot hverandre og differansen mellom punktene går mot null. 

stigningstallet til $k$ til sekanten gjennom punktet $a$ og $a+∆a$ kan skrives slik:

$$k = \space  \frac{f(a+∆a)-f(xa)}{∆a}$$

fordi $\frac{∆y}{∆x}$ gir oss stigningstallet mellom to punkter, og hvis vi ser på utrykket for k så er det akkurat samme greia, differansen på $y$-aksen delt på differansen på    $x$-aksen gir oss stigningstallet.

#### b) 

En tanget er en linje som berører en graf i bare et punkt og linjen har samme stigningstall som grafen i punktet.

Stigningstallet til $k$ til tangenten gjennom punktet $a$ på funksjonen $f$ kan skrives slik:

$$f'(a) =\space \lim_{∆a\rightarrow0} \space \frac{f(a+∆a)-f(a)}{∆a}$$

fordi $\frac{∆y}{∆x}$ gir oss stigningstallet mellom to punkter, men hvis vi sier at differansen mellom de to punktene på x-aksen blir veldig liten, altså, $∆a$ går mot null, da blir disse to punktene så nære hverandre at vi kan se det som et punkt, og vi får et uttrykk for stigningstallet.

#### c)

Derivasjon er oppfatningen av endring mellom to punkter. 

#### d)

Newtons kvotient er:

Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.

$$f'(x) ≈ \frac{f(x+h)-f(x)}{h}$$

Definisjonen av den deriverte er:

$$f'(a) =\space \lim_{∆a\rightarrow0} \space \frac{f(a+∆a)-f(a)}{∆a}$$


Forskjellen mellom newtons kvotient og definisjonen av den deriverte er at med newtons kvotient må h være et tall nære null, mens i definisjonen av den deriverte kan vi la $∆a$ gå mot et tall uendelig nærme null.

```python
from pylab import *

def fder(f,x,dx=1E-8):
    return (f(x+dx)-f(x))/dx

def f(x):
    return x**2

df = fder(f,1)

print(df)
```

```python
from pylab import *

def fder(f,x,dx):
    return (f(x+dx)-f(x))/dx

def f(x):
    return x**2

analytisk_verdi = 2

dx = []
feilen = []

for i in range(1,17):
    DX = 10**(i*(-1))
    F = abs(analytisk_verdi-fder(f,1,DX))
    dx.append(log10(DX))
    feilen.append(log10(F))
   

plot(dx,feilen)
xlabel("∆x i logarytmisk skala")
ylabel("feilen i logarytmisk skala")
show()
```
